# cohort plots

```python
dfs = df[df.lifetime < 10]
dfsm = df_mutant[df_mutant.lifetime < 10]

def scale(d, scale=56):
    for i in range(d.shape[0]):
        d['Y'].iloc[i] = [y/scale for y in d['Y'].iloc[i]]
    return d

dfs = scale(dfs)
dfsm = scale(dfsm)

for i in range(1000):
    plt.plot(dfs.iloc[i]['Y'], alpha=0.25, color='gray')
    plt.plot(dfsm.iloc[i]['Y'], alpha=0.25, color='blue')
plt.xlabel('Time')
plt.ylabel('vps4 Amplitude')

plt.ylim((0, 60))
```
